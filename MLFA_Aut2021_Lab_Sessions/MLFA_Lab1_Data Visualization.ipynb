# Introduction to Data Visualization 
### Types of plots: Line, Scatter, Histogram, Box, Bar, Pie
### Library for visulaization: matplotlib 

```python
#The the library that we will use is matplotlib [Other libraries- Seaborn, ggplot]
#We will import only pyplot sub module
#pyplot is collection of command style functions that make matplotlib work like MATLAB
import matplotlib.pyplot as plt
```

### Generate some synthetic data
We will not load any actual data for this tutorial. 


```python
# Example 1 - Draw a line in a diagram from position (0,0) to position (5,25)
import numpy as np
x = np.array([0, 5])
y = np.array([0, 25])

# Plotting
plt.plot(x, y)  # takes parameters for specifying points in the diagram
plt.show()
# By default, plot() draws a line from point to point.
```

```python
# Use the keyword "marker" to emphasize each point with a specified marker.
# Make it more readable by setting the axis names and a plot title
plt.plot(x,y, marker='o')
plt.xlabel("x")
plt.ylabel("y")
plt.title("Line plotting")
plt.show()
```

```python
# Plot the predicted data with some more customization
plt.plot(x,y, color='g', marker='o', linestyle='dotted', linewidth=2, markersize=10)
# Different colors can be assigned such as r, g, b, y, k. 
# Different types of markers can be used such as *, x, + ans so on
# Linestyle can be solid, dashed, dotted or dashed-dotted.
# Linewidth and markersize can also be adjusted 
plt.xlabel("x", color='blue', size = 10)
plt.ylabel("y", color='blue', size= 10)
plt.title("Line plotting with customization", color='red', size= 20)
#plt.grid()
# plt.grid(color = 'blue', linestyle = 'dashed', linewidth = 0.5)
plt.show()
```

### Display multiple plots in one figure
Using subplot


```python
#plot 1:
x1 = np.array([0, 1, 2, 3])
y1 = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1)
plt.plot(x1,y1)
plt.title('Plot1')

#plot 2:
x2 = np.array([0, 1, 2, 3])
y2 = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)
plt.plot(x2,y2)
plt.title('Plot2')

plt.suptitle('Displaying plots side by side')
plt.show()
```

Example 2 - Create synthetic rain fall for 30 days data using random number generator and plot it.

```python
# We need 31 days(x) and rainfall on each day in mm (y_actual)
# Using numpy to generate the random numbers

np.random.seed(64) # We use this to get same data at evry run

x = range(1,32) 
y_actual = [np.random.uniform(0, 5) for _ in range(1,32)] #Represent actual rainfall amount; random uniform distribution b/w 0 and 5 with 31 samples

# Lets plot and make it more readable by setting the axis names and a plot title
plt.plot(x, y_actual)
plt.xlabel("Day")
plt.ylabel("mm Rainfall")
plt.title("Daily Rainfall Aug 2020")
plt.grid()
plt.show()
```

```python
#Supposed you developed ML algo that predicts the rainfall. Lets call the data as y_pred 
y_pred = [y + np.random.uniform(-2,2) for y in y_actual]
```

```python
#Plot the predicted data with some more customization
plt.plot(y_pred, color='r', marker='o', linestyle='dashed', linewidth=2, markersize=12)
plt.xlabel("Day")
plt.ylabel("mm Rainfall Predicted")
plt.title("Predicted Daily Rainfall Aug 2020")
plt.show()
```

```python
# To see the plots side by side
# plt.subplot() helps us to plot multiple figures in a grid
plt.subplot(1,2,1)
plt.xlabel("Day")
plt.ylabel("mm Rainfall")
plt.title("Daily Rainfall Aug 2020")
plt.plot(x, y_actual, 'ro-')

plt.subplot(1,2,2)
plt.xlabel("Day")
plt.ylabel("mm Rainfall Predicted")
plt.title("Predicted Daily Rainfall Aug 2020")
plt.plot(x, y_pred, 'bx-')
plt.show()
```

```python
#We can see above that the plots labels are overlapping
#It is diificult to control the plots in command style 
#A better way is to use object oriented coding style 
```

```python
#Plottig figure in object oriented way
fig = plt.figure() # an empty figure with no axes
axes = fig.add_axes([0.0, 0.0, 1, 1]) # add axes to the fig
axes.plot(x, y_actual, 'gx-')
axes.set_xlabel('Day')
axes.set_ylabel('mm Rainfall')
axes.set_title('Daily Rainfall Aug 2020')
plt.show()
```

```python
#We can do plot inside plot
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1]) #main axes
ax1.set_xlabel('Day')
ax1.set_ylabel('mm Rainfall')
ax1.set_title('Daily Rainfall Aug 2020')

ax2 = fig.add_axes([0.7, 0.7, 0.2, 0.2]) #inset axes
#ax2.set_xlabel('Day')
#ax2.set_ylabel('mm Rainfall Predicted')
ax2.set_title('Predicted')

ax1.plot(x, y_actual, 'r')
ax2.plot(x, y_pred, 'b')
plt.show()
```

```python
#Another important method is plt.subplots [its different from subplot that you see earlier]
fig, ax = plt.subplots(1,2) #Grid size

ax[0].set_xlabel('Day')
ax[0].set_ylabel('mm Rainfall')
ax[0].set_title('Daily Rainfall Aug 2020')
ax[0].plot(x, y_actual, 'b')

ax[1].set_xlabel('Day')
ax[1].set_ylabel('mm Rainfall Predicted')
ax[1].set_title('Predicted Daily Rainfall Aug 2020')
ax[1].plot(x, y_pred, 'r')


plt.show()
```

```python
#Solves the overlapping problem easily with fig.tight_layout()
fig, ax = plt.subplots(1,2) #Grid size

ax[0].set_xlabel('Day')
ax[0].set_ylabel('mm Rainfall')
ax[0].set_title('Daily Rainfall Aug 2020')
ax[0].plot(x, y_actual, 'b')

ax[1].set_xlabel('Day')
ax[1].set_ylabel('mm Rainfall Predicted')
ax[1].set_title('Predicted Daily Rainfall Aug 2020')
ax[1].plot(x, y_pred, 'r')

fig.tight_layout()
plt.show()
```

```python
#We can use for loop
fig, ax = plt.subplots(1,2)
col = ['r', 'g']
data = [y_actual, y_pred]
for i, axes in enumerate(ax):
    axes.plot(x, data[i], col[i])

fig.tight_layout()
plt.show()
```

```python
#We can control the size of the figure, resolution (dpi=dots per inch) and many more
fig, ax = plt.subplots(figsize = (8, 4), dpi = 100)
ax.plot(x, y_actual, 'r')
ax1.set_xlabel('Day')
ax1.set_ylabel('mm Rainfall')
ax1.set_title('Daily Rainfall Aug 2020')
plt.show()
```

```python
#Saving a figure in folder
fig.savefig('rainfall_plot.png', dpi = 100)
```

```python
#We can draw multiple plots in the same canvas
fig, ax = plt.subplots( dpi = 100)
ax.plot(x, y_actual, 'r', label = 'Actual')
ax.plot(x, y_pred, 'b', label = 'Predicted')
ax.set_xlabel('Day')
ax.set_ylabel('mm Rainfall')
ax.legend(loc=5)
ax.set_title('Actual and Predicted Daily Rainfall Aug 2020')
plt.show()
```

```python
#Hexadecimal color values
fig, ax = plt.subplots()
ax.plot(x, y_actual, markersize = 12, linewidth=3, color = '#005454')
plt.show()
```

```python
#Limiting the axis
fig, ax = plt.subplots(1,3, figsize = (12, 4))
ax[0].plot(x, y_actual)
ax[1].plot(x, y_pred)
ax[2].plot(x, np.cumsum(y_actual)) # Cumulative rainfall
ax[2].set_xlim(0, 15) #Limit view to 15 days
plt.show()
```

### Other types of plots

```python
#Scatter plot - Just the points
#Useful when checking relation between two variables - such as temp and rainfall
plt.scatter(x, y_actual)
plt.show()
```

```python
# two plots on the same figure
plt.scatter(x, y_actual, color='pink')
plt.scatter(x, y_pred, color='yellow' )
plt.show()
```

```python
#Histogram 
plt.hist(y_actual, bins=[0,1,2,3,4], rwidth=0.9)
plt.show()
```

```python
#Boxplot 
data = [y_actual, y_pred]
plt.boxplot(data, vert = True, patch_artist=True)
plt.show()
```
